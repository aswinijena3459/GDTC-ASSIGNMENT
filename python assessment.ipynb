## Task #8
Mobile & Travel Insurance premium are discounted by 10%

Health and Property Insurance premium are increased by 7%

Life and Motor Insurance premium are marginally increased by 2%

What will be overall change in % of the Premium Amount Collected for all these Customer?

```python
insurance_df['INSURANCE_TYPE'].unique()
```

```python
premium=pd.DataFrame(insurance_df.groupby('INSURANCE_TYPE')['PREMIUM_AMOUNT'].sum())
premium
```

```python
premium_df['PREMIUM_AMOUNT'].sum()
```

```python
def new_premium(premium_df):
    if (premium_df['INSURANCE_TYPE']=='Health') | (premium_df['INSURANCE_TYPE']=='Property'): 
        val = (premium_df['PREMIUM_AMOUNT'] * 0.07) + premium_df['PREMIUM_AMOUNT']
    elif (premium_df['INSURANCE_TYPE']=='Mobile') | (premium_df['INSURANCE_TYPE']=='Travel'): 
        val =  premium_df['PREMIUM_AMOUNT'] - (premium_df['PREMIUM_AMOUNT'] * 0.10) 
    elif (premium_df['INSURANCE_TYPE']=='Life') | (premium_df['INSURANCE_TYPE']=='Motor'): 
        val =  premium_df['PREMIUM_AMOUNT'] + (premium_df['PREMIUM_AMOUNT'] * 0.02) 
    return val

premium_df['NEW_PREMIUM']=premium_df.apply(new_premium,axis=1)
```

```python
premium_df
```

```python
premium_df['PERCENTAGE_CHANGE_PREMIUM']=((premium_df['NEW_PREMIUM']-premium_df['PREMIUM_AMOUNT'])/premium_df['NEW_PREMIUM'])*100
premium_df
```

## Task #9
Business wants to give discount to customer who are loyal and under stress due to Covid19.
They have laid down an eligibility Criteria as follow

IF CUSTOMER_TENURE > 60 AND EMPLOYMENT_STATUS = “N”
AND NO_OF_FAMILY_MEMBERS >=4 THEN 1 ELSE 0

Create a new column “ELIGIBLE_FOR_DISCOUNT” and find it
mean.

```python
def discount(insurance_df):
    if (insurance_df['TENURE'] > 60) & (insurance_df['EMPLOYMENT_STATUS'] == 'N') & (insurance_df['NO_OF_FAMILY_MEMBERS'] >= 4):
        val=1
    else:
        val=0
    return val
insurance_df['ELIGIBLE_FOR_DISCOUNT'] = insurance_df.apply(discount, axis=1)
```

```python
insurance_df['ELIGIBLE_FOR_DISCOUNT'].mean()
```

## Task #12
Business wants to find all Suspicious Employees (Agents).
IF TOTAL CLAIM AMOUNT which meet below criteria is >= 15000 THEN AGENT is classified
as Suspicious ELSE Not
CLAIM_STATUS = Approved AND CUSTOMER_RISK_SEGMENTATION = High
AND INCIDENT_SEVERITY = “Major Loss”
If Suspicious, then 1 ELSE 0. Find mean of this column

```python
suspicion=insurance_df[['CLAIM_STATUS','RISK_SEGMENTATION','INCIDENT_SEVERITY','CLAIM_AMOUNT']][(insurance_df['CLAIM_STATUS']=='A' )&( insurance_df['RISK_SEGMENTATION']=='H') & (insurance_df['INCIDENT_SEVERITY']=='Major Loss')]
suspicion
```

```python
insurance_df.columns
```

```python
def suspicious(suspicion):
    if (suspicion['CLAIM_AMOUNT'] >=15000) :
        val=1
    else:
        val=0
    return val

suspicion['SUSPICIOUS']=suspicion.apply(suspicious,axis=1)
```
